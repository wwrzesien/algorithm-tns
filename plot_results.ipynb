```python
import matplotlib.pyplot as plt
import pickle
```

```python
with open("results.pickle", "rb") as file:
    data = pickle.load(file)
```

```python
def plot_runtime_minsupp(data={"comb": [1,2,3], "1000": [3,4,5], "1500": [1,2,5], "supp": [1, 4, 10]}):
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, constrained_layout=True)
    ax1.plot(data["supp"], data["comb"], 'tab:gray')
    ax1.set_title('Zbi贸r twitter_comb')
    ax2.plot(data["supp"], data["1000"], 'tab:gray')
    ax2.set_title('Zbi贸r twitter_1000')
    ax3.plot(data["supp"], data["1500"], 'tab:gray')
    ax3.set_title('Zbi贸r twitter_1500')
    ax4.plot(data["supp"], data["1500"], 'tab:gray')
    ax4.set_title('Zbi贸r twitter_2000')

    for ax in fig.get_axes():
        ax.set(xlabel='Min. wsparcie', ylabel='Czas wykonania (s)')
        # ax.label_outer()
```

```python
plot_runtime_minsupp(data)
```
